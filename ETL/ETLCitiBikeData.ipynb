```python
# Importing dependencies
import pandas as pd
import numpy as np
import datetime as dt
pd.set_option('max_colwidth', 400)
```

```python
# Reading 2019 data into a Pandas DataFrame
June2019 = pd.read_csv('201906-citibike-tripdata.csv')
July2019 = pd.read_csv('201907-citibike-tripdata.csv')
August2019 = pd.read_csv('201908-citibike-tripdata.csv')
September2019 = pd.read_csv('201909-citibike-tripdata.csv')

# Reading 2020 data into a Pandas DataFrame
June2020 = pd.read_csv('202006-citibike-tripdata.csv')
July2020 = pd.read_csv('202007-citibike-tripdata.csv')
August2020 = pd.read_csv('202008-citibike-tripdata.csv')
September2020 = pd.read_csv('202009-citibike-tripdata.csv')

# Concatenating 2019 data into a single DataFrame
frames2019 = [June2019, July2019, August2019, September2019]
df2019 = pd.concat(frames2019)
df2019.head()

# Concatenating 2020 data into a single DataFrame
frames2020 = [June2020, July2020, August2020, September2020]
df2020 = pd.concat(frames2020)
df2020.head()

# Concatenating 2019 and 2020 data into a single DataFrame
frames20192020 = [df2019, df2020]
df_20192020 = pd.concat(frames20192020)
df_20192020.head()
```

```python
# Getting a brief summary of the new DataFrame
df_20192020.info()
```

```python
# changing gender names to male and female
df_20192020["gender"].replace({1: 'Male', 2: 'Female', 0: 'Unknown'}, inplace=True)

# breaking out the year, month, day, and hour from the starttime column and adding them as new columns
df_20192020['Trip Year'] = df_20192020.starttime.str.split('-', expand=True)[0] 
df_20192020['Trip Month'] = df_20192020.starttime.str.split('-', expand=True)[1]
df_20192020['Trip Day'] = df_20192020.starttime.str.split('-', expand=True)[2].str.split(' ', expand=True)[0]
df_20192020['Trip Hour'] = df_20192020.starttime.str.split(' ', expand=True)[1].str.split(':', expand=True)[0]
df_20192020.head() 
```

```python
# giving names to months
df_20192020["Trip Month"].replace({'06': 'June', '07': 'July', '08': 'August', '09': 'September'}, inplace=True)
#checking last 10 rows
df_20192020.tail()
```

```python
# Checking the data types of the new columns
df_20192020.dtypes
```

```python
# Function to convert seconds to "HH:MM:SS" format
def convert_seconds_to_hh_mm_ss(seconds):
    m, s = divmod(seconds, 60)
    h, m = divmod(m, 60)
    return "{:02}:{:02}:{:02}".format(int(h), int(m), int(s))

# Assuming df_20192020 is your DataFrame with a 'tripduration' column
df_20192020['NewTripDuration'] = df_20192020['tripduration'].apply(lambda x: convert_seconds_to_hh_mm_ss(x))

# Remove the old 'tripduration' column
df_20192020.drop('tripduration', axis=1, inplace=True)

# Display the updated DataFrame
df_20192020.head()

```

```python
# checking column names as a list
df_20192020.columns.tolist()
```

```python
# Converting the time part to '00:00:00' format for 'starttime' and 'stoptime' columns
df_20192020[['starttime', 'stoptime']] = df_20192020[['starttime', 'stoptime']].apply(lambda col: pd.to_datetime(col).dt.strftime('%Y-%m-%d %H:%M:%S').str.replace(r' \d{2}:\d{2}:\d{2}$', ' 00:00:00'))

# Rename columns with best possible names
df_20192020 = df_20192020.rename(columns={
    'start station id': 'StartStationID',
    'start station name': 'StartStationName',
    'start station latitude': 'StartStationLatitude',
    'start station longitude': 'StartStationLongitude',
    'end station id': 'EndStationID',
    'end station name': 'EndStationName',
    'end station latitude': 'EndStationLatitude',
    'end station longitude': 'EndStationLongitude',
    'bikeid': 'BikeID',
    'usertype': 'UserType',
    'birth year': 'BirthYear',
    'gender': 'Gender',
    'Trip Year': 'TripYear',
    'Trip Month': 'TripMonth',
    'Trip Day': 'TripDay',
    'Trip Hour': 'TripHour',
    'NewTripDuration': 'TripDuration'
})

# Displaying the updated DataFrame
df_20192020.head()

```

```python
# checking column names as a list
df_20192020.columns.tolist()
```

```python
# checking the type of each column
df_20192020.dtypes
```

```python
# Converting 'TripYear' to int64
df_20192020['TripYear'] = pd.to_numeric(df_20192020['TripYear'], errors='coerce').astype('Int64')

# Define the new column order
new_column_order = ['UserType', 'BirthYear', 'Gender', 'TripYear', 'TripMonth', 'TripDay', 'TripHour',
                    'starttime', 'stoptime', 'TripDuration', 'StartStationID', 'StartStationName',
                    'StartStationLatitude', 'StartStationLongitude', 'EndStationID',
                    'EndStationName', 'EndStationLatitude', 'EndStationLongitude', 'BikeID']

# Reorganize the columns
df_20192020 = df_20192020[new_column_order]

# Calculate the 'Age' column
df_20192020['Age'] = df_20192020['TripYear'] - df_20192020['BirthYear']

# Display the updated DataFrame
df_20192020.head()
```

```python
# Defining the new column order
new_column_order = ['BikeID', 'UserType', 'BirthYear', 'Gender', 'Age', 'TripYear', 'TripMonth', 'TripDay', 'TripHour',
                    'starttime', 'stoptime', 'TripDuration', 'StartStationID', 'StartStationName',
                    'StartStationLatitude', 'StartStationLongitude', 'EndStationID',
                    'EndStationName', 'EndStationLatitude', 'EndStationLongitude']

# Reorganizing the columns
df_20192020 = df_20192020[new_column_order]

# Display the updated DataFrame
df_20192020.head()
```
