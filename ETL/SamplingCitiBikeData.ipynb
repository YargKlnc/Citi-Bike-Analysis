##### data frame too large for Tableau, randomly sampling from 18 million rows to 14 million

```python
# Importing dependencies
import pandas as pd
pd.set_option('max_colwidth', 400)
```

```python
# Reading 2019 data into a Pandas DataFrame
CitiBike_df = pd.read_csv('citibike.csv')

# reading head of data
CitiBike_df.head()
```

```python
# randomly sampling 14 million rows from the data
df_reduced = CitiBike_df.sample(n=14000000)
```

```python
# Exporting the DataFrame as a CSV file inside the same folder
df_reduced.to_csv("citibikeshortened.csv", index=False)
```
